```python
import pandas as pd
data = {
    '나이':[22,35,14,55,42]
}

df = pd.DataFrame(data)

def categorize(data):
    if 0 < data <= 10:
        return "10대"
    elif 10 < data <= 20:
        return "20대"
    elif 20 < data <= 30:
        return "30대"
    else:
        return "30대 이상"

df['검사나이'] = df['나이'].apply(categorize)
    
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "문항1": ['1', '2', "", '2'],
    "문항2": ["", '2', '3', ""]
})

# 문자열 합치기
df["문항"] = df["문항1"].fillna("") + "," + df["문항2"].fillna("")

# 앞뒤 콤마 제거
df["문항"] = df["문항"].str.rstrip(",")
df["문항"] = df["문항"].str.lstrip(",")

# 기존 컬럼 삭제
df = df.drop(columns=["문항1", "문항2"])

print(df)
```

```python
import pandas as pd

df = pd.DataFrame({
    "문항1": ['1', '2', "", '2'],
    "문항2": ["", '2', '3', ""],
    "문항3": ["", '1', '2', ""]
})

cols = ["문항1", "문항2", "문항3"]

# 빈 문자열 -> 무응답
df[cols] = df[cols].replace(r'^\s*$', '무응답', regex=True)

# 문자열 합치기
df["문항"] = (
    df["문항1"].fillna("") + "," +
    df["문항2"].fillna("") + "," +
    df["문항3"].fillna("")
)

print(df)
```
